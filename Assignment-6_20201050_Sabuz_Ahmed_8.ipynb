```python
#Task-2
class Assassin:
    totalAssasinCount = 0

    def __init__(self, name, su_Rate):
        self.name = name
        self.successRate = su_Rate
        Assassin.totalAssasinCount += 1

    @classmethod
    def failureRate(cls, name, fail_Count):
        return cls(name, 100 - fail_Count)
        Assassin.totalAssasinCount += 1

    @classmethod
    def failurePercentage(cls, name, fail_Percentage):
        return cls(name, 100 - fail_Percentage)
        Assassin.totalAssasinCount += 1

    def printDetails(self):
        print("Name: {}".format(self.name))
        print("Success rate: {}%".format(self.successRate))
        print("Total number of Assassin: {}".format(self.totalAssasinCount))


john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
#Task-4
class Travel:
    count = 0

    def __init__(self, source, desti):
        self.src = source
        self.des = desti
        self.time = 1
        Travel.count += 1

    def set_time(self, time=''):
        if time == '':
            self.time = time
        else:
            self.time += time - 1

    def set_source(self, chr):
        self.src = chr

    def set_destination(self, changed):
        self.des = changed

    def display_travel_info(self):
        return (f'Source:{self.src}\nDestination:{self.des}\nFlight Time:{self.time}:00')


print('No. of Traveller =', Travel.count)
print("=======================")
t1 = Travel("Dhaka", "India")
print(t1.display_travel_info())
print("=======================")
t2 = Travel("Kuala Lampur", "Dhaka")
t2.set_time(23)
print(t2.display_travel_info())
print("=======================")
t3 = Travel("Dhaka", "New_Zealand")
t3.set_time(15)
t3.set_destination("Germany")
print(t3.display_travel_info())
print("=======================")
t4 = Travel("Dhaka", "India")
t4.set_time(9)
t4.set_source("Malaysia")
t4.set_destination("Canada")
print(t4.display_travel_info())
print("=======================")
print('No. of Traveller =', Travel.count)
```

```python
#Task-5
import datetime

now = datetime.datetime.now()


class Employee:
    name = None
    workingPeriod = 0

    def __init__(self, employee_Name, employee_Experience):
        self.name = employee_Name
        self.workingPeriod = employee_Experience

    def employeeByJoiningYear(employeeName, employee_Year):
        yearOfExperience = now.year - employee_Year
        return Employee(employeeName, yearOfExperience)

    def experienceCheck(yearOfExperience, gender):
        noun = "He" if gender == "male" else "She"
        experience = "experienced" if yearOfExperience > 2 else "not experienced"
        return f"{noun} is {experience}"


employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, 'male'))
print(Employee.experienceCheck(3, 'female'))
```

```python
#TASK-7

class Cat:
    Number_of_cats = 0
    def __init__(self, colour='WHITE', Position='Sitting'):
        self.colour = colour
        self.Position = Position
        Cat.Number_of_cats = Cat.Number_of_cats + 1
    @classmethod
    def no_parameter(class0):
        colour = 'White'
        Position = 'sitting'
        obj = class0(colour,Position)
        return obj
    @classmethod
    def first_parameter(class1,colour):
        class1.colour = colour
        Position = 'sitting'
        obj =class1(colour,Position)
        return obj
    @classmethod
    def second_parameter(class2, Position):
        colour = 'White'
        class2.Position = Position
        obj = class2(colour, Position)
        return obj
    def changeColor(self,colour):
        self.colour = colour
    def printCat(self):
        print(f'{self.colour} cat is {self.Position}')

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#TASK-8
import math
class Cylinder:

    radius =3
    height =8
    def __init__(self,Radius,Height):
        self.radius = Radius
        self.height = Height
        print(f'Default radius={Cylinder.radius} and height={Cylinder.height}.')
        Cylinder.radius = Radius
        Cylinder.height = Height
        print(f'Updated: radius={Cylinder.radius} and height={Cylinder.height}.')
    @classmethod
    def swap(class0,Radius,Height):
        obj =class0(Height,Radius)
        return obj
    @classmethod
    def changeFormat(class1,total):
        nRad,nHeight = total.split('-')
        obj1 = class1 (float(nRad),float(nHeight))
        return obj1
    @staticmethod
    def volume(Radius,Height):
        print('Volume:',(math.pi) * Radius * Radius * Height)
    @staticmethod
    def area(Radius,Height):
        print('Area:',(2*math.pi*Radius*Radius) + (2*math.pi*Radius*Height))

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)


```

```python
#Task-10

class SultansDine:
    numBranch = 0
    totalSell = 0
    branches = []

    def __init__(self, branchName):
        self.branchName = branchName
        self.sellAmount = 0

        SultansDine.numBranch += 1
        SultansDine.branches.append(self)

    def sellQuantity(self, quantity):
        if quantity < 10:
            amount = quantity * 300
        elif quantity < 20:
            amount = quantity * 350
        else:
            amount = quantity * 400

        self.sellAmount += amount
        SultansDine.totalSell += amount

    def branchInformation(self):
        print("Branch Name: {}".format(self.branchName))
        print("Branch Sell: {} Taka".format(self.sellAmount))

    @staticmethod
    def details():

        print("Total Number of branch(s): {}".format(SultansDine.numBranch))
        print("Total Sell: {} Taka".format(SultansDine.totalSell))

        for branch in SultansDine.branches:
            print("Branch Name: {}, Branch Sell: {} Taka".format(branch.branchName, branch.sellAmount))
            print("Branch consists of total sell's: {:.2f} %%".format(branch.sellAmount * 100 / SultansDine.totalSell))



SultansDine.details()
print("#################################")
dhanmodi = SultansDine("Dhanmodi")
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print("----------------------------------")
SultansDine.details()

print("==================================")

baily_road = SultansDine("Baily Road")
baily_road.sellQuantity(15)
baily_road.branchInformation()
print("----------------------------------")
SultansDine.details()

print("==================================")
gulshan = SultansDine("Gulshan")
gulshan.sellQuantity(9)
gulshan.branchInformation()
print("----------------------------------")
SultansDine.details()
```
